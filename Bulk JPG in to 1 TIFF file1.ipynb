```python
# pip install Pillow  # if not install

from PIL import Image
import glob

# Step 1: Get a list of all .jpg files in the directory
jpg_files = glob.glob("path_to_jpg_images/*.jpg")  # Change to your folder path

# Step 2: Open the first image as the base
first_image = Image.open(jpg_files[0])

# Step 3: Open the remaining images
image_list = []
for file in jpg_files[1:]:
    img = Image.open(file)
    image_list.append(img)

# Step 4: Save the images as a multi-page TIFF file
first_image.save("output_images/output_file.tiff", save_all=True, append_images=image_list)

print("All images are successfully combined into a multi-page TIFF file!")
```
